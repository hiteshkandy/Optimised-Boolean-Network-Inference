```python

# ==============================================================
# FULL PIPELINE: load models  compute ALL scores (functional + network + motifs)
#                build correlation matrices (+ optional heatmaps)
#
# FIXES FOR YOUR CURRENT FAILURE:
#   In your copy, compute_all_FFLs() returns ffl_res[3] with shape (N,) (not (8,N)).
#   That means it is a TOTAL FFL count per network (or similar), not "per-type".
#
#   We therefore compute expected FFL counts per type ourselves using the SAME null
#   formula used inside analyse_database13.get_null_expectations_ffl:
#       expected_type[t,i] = total_ffls[i] * p[i]^npos[t] * (1-p[i])^(3-npos[t])
#   where p[i] = prop_pos_separate2 (from get_null_expectations_fbl)
#
#   This avoids relying on a mismatched get_null_expectations_ffl signature and
#   removes the (8,N) requirement for nr_specific_ffls.
# ==============================================================
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools

import analyse_database13 as ad
import canalizing_function_toolbox_v16 as can

# -----------------------------
# Helper utilities
# -----------------------------
def safe_div(a, b):
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    out = np.full_like(a, np.nan, dtype=float)
    m = (b != 0) & ~np.isnan(a) & ~np.isnan(b)
    out[m] = a[m] / b[m]
    return out

def safe_nanmean_list(x):
    vals = []
    def _add(item):
        if item is None:
            return
        if isinstance(item, (np.generic,)):
            try:
                vals.append(float(item))
            except Exception:
                pass
            return
        if isinstance(item, (int, float, np.integer, np.floating)):
            vals.append(float(item))
            return
        if isinstance(item, np.ndarray):
            if item.dtype == object:
                for sub in item.ravel():
                    _add(sub)
            else:
                try:
                    arr = item.astype(float, copy=False).ravel()
                    vals.extend(arr.tolist())
                except Exception:
                    for sub in item.ravel():
                        _add(sub)
            return
        if isinstance(item, (list, tuple)):
            for sub in item:
                _add(sub)
            return
        try:
            vals.append(float(item))
        except Exception:
            return
    _add(x)
    if len(vals) == 0:
        return np.nan
    arr = np.asarray(vals, dtype=float)
    if arr.size == 0:
        return np.nan
    return np.nanmean(arr)

def derrida_values_safe(Fs, Is, degrees, max_degree, nsim=2000):
    Nloc = len(Fs)
    valid = []
    for i in range(Nloc):
        F = Fs[i]
        ok = True
        for f in F:
            if f is None:
                ok = False
                break
            try:
                if len(f) == 0:
                    ok = False
                    break
            except TypeError:
                ok = False
                break
        if ok:
            valid.append(i)

    derr = np.full(Nloc, np.nan, dtype=float)
    if len(valid) == 0:
        return derr

    Fs_valid  = [Fs[i] for i in valid]
    Is_valid  = [Is[i] for i in valid]
    deg_valid = [degrees[i] for i in valid]

    res = ad.get_derrida_values(Fs_valid, Is_valid, deg_valid, max_degree=max_degree, nsim=nsim)
    derr_valid = res[0] if (isinstance(res, tuple) and len(res) >= 1) else res

    derr_valid = np.asarray(derr_valid, dtype=float)
    for idx, val in zip(valid, derr_valid):
        derr[idx] = val
    return derr

def expected_ffl_per_type_from_total(total_ffls, p_pos):
    """
    Implements the same null model core used in analyse_database13.get_null_expectations_ffl:
      expected_number_per_type[i,t] = total_ffls[i] * p_pos[i]^npos[t] * (1-p_pos[i])^(3-npos[t])
    Returns array shape (8, N).
    """
    total_ffls = np.asarray(total_ffls, dtype=float).reshape(-1)  # (N,)
    p_pos      = np.asarray(p_pos, dtype=float).reshape(-1)       # (N,)
    Nloc = len(total_ffls)

    LEGEND = list(itertools.product([0, 1], repeat=3))  # 8 sign-pattern classes
    npos = np.array([sum(bits) for bits in LEGEND], dtype=int)    # (8,)

    out = np.full((8, Nloc), np.nan, dtype=float)
    for t in range(8):
        out[t, :] = total_ffls * (p_pos ** npos[t]) * ((1.0 - p_pos) ** (3 - npos[t]))
    return out

# ==============================================================
# 1) LOAD MODELS
# ==============================================================
max_degree = 12
max_N = 10_000

out = ad.load_models_included_in_meta_analysis(max_degree=max_degree, max_N=max_N)

Fs                = out[0]
Is                = out[1]
degrees           = out[2]
degrees_essential = out[3]
variabless        = out[4]
constantss        = out[5]
models_loaded     = out[6]
models_not_loaded = out[7]
similar_sets      = out[8]
extra_bookkeeping = out[9]

n_variables       = np.asarray(out[10], dtype=int)
n_constants       = np.asarray(out[11], dtype=int)
max_degree_used   = int(out[12])

N = len(Fs)
print("=== Meta-analysis model set loaded ===")
print(f"Number of models (N): {N}")
print(f"Max in-degree used:   {max_degree_used}")
print(f"Models not loaded:    {len(models_not_loaded)}")
print()

# ==============================================================
# 2) SUBSET TO GENE NODES + BUILD ESSENTIAL FUNCTION LISTS
# ==============================================================
Fs_genes              = [Fs[i][:n_variables[i]] for i in range(N)]
degrees_genes         = [degrees[i][:n_variables[i]] for i in range(N)]
degrees_essential_gen = [degrees_essential[i][:n_variables[i]] for i in range(N)]

Fs_essential = [
    [f for f, ke in zip(Fs_genes[i], degrees_essential_gen[i])
     if (ke > 0 and (f is not None) and (len(f) > 0))]
    for i in range(N)
]

# ==============================================================
# 3) CANALIZING DEPTHS + NODE-WISE CANALIZATION FRACTIONS (safe)
# ==============================================================
depths = [np.full(n_variables[i], np.nan, dtype=float) for i in range(N)]

for m in range(N):
    for j in range(n_variables[m]):
        f  = Fs_genes[m][j]
        ke = degrees_essential_gen[m][j]
        if ke <= 0 or f is None or len(f) == 0:
            continue
        if len(f) != 2**int(ke):
            continue
        f = np.asarray(f, dtype=int)
        try:
            (_, depth, _, _, _) = can.get_canalizing_depth_inputs_outputs_corefunction(f)
            depths[m][j] = float(depth)
        except Exception:
            depths[m][j] = np.nan

mean_depth = np.full(N, np.nan, dtype=float)
prop_ncf   = np.full(N, np.nan, dtype=float)

for m in range(N):
    d_arr  = np.asarray(depths[m], dtype=float)
    ke_arr = np.asarray(degrees_essential_gen[m], dtype=float)
    mask = (ke_arr > 0) & (~np.isnan(d_arr))
    if mask.sum() == 0:
        continue
    mean_depth[m] = np.nanmean(d_arr[mask])
    prop_ncf[m]   = np.nanmean(d_arr[mask] / ke_arr[mask])

# ==============================================================
# 4) CANALIZING STRENGTHS  mean per network (robust)
# ==============================================================
can_strengths = ad.get_canalizing_strengths(Fs_essential)

def _strength_payload(cs):
    return cs[0] if (isinstance(cs, tuple) and len(cs) >= 1) else cs

mean_can_strength = np.array(
    [safe_nanmean_list(_strength_payload(cs)) for cs in can_strengths],
    dtype=float
)

# ==============================================================
# 5) INPUT REDUNDANCY, EFFECTIVE CONNECTIVITY, BIAS p(1p), Ke, cov(Ke,bias)
# ==============================================================
bias_list = [[] for _ in range(N)]
IR_list   = [[] for _ in range(N)]
EC_list   = [[] for _ in range(N)]
Ke_list   = [[] for _ in range(N)]

for m in range(N):
    for f, ke in zip(Fs_genes[m], degrees_essential_gen[m]):
        d = int(ke)
        if d <= 0 or f is None or len(f) == 0:
            continue
        if len(f) != 2**d:
            continue

        f = np.asarray(f, dtype=int)
        p1 = float(np.sum(f)) / float(2**d)
        bias_list[m].append(p1 * (1.0 - p1))

        if d <= 8:
            try:
                IR_list[m].append(can.get_input_redundancy(f, n=d))
            except Exception:
                IR_list[m].append(np.nan)
        else:
            IR_list[m].append(np.nan)

        if d <= max_degree_used:
            try:
                effs = can.get_edge_effectiveness(f, n=d)
                EC_list[m].append(np.nansum(effs))
            except Exception:
                EC_list[m].append(np.nan)
        else:
            EC_list[m].append(np.nan)

        Ke_list[m].append(float(d))

mean_bias = np.array([safe_nanmean_list(b)  for b  in bias_list], dtype=float)
mean_IR   = np.array([safe_nanmean_list(ir) for ir in IR_list],   dtype=float)
mean_EC   = np.array([safe_nanmean_list(ec) for ec in EC_list],   dtype=float)
mean_Ke   = np.array([safe_nanmean_list(ke) for ke in Ke_list],   dtype=float)

cov_Ke_bias = np.full(N, np.nan, dtype=float)
for m in range(N):
    k = np.asarray(Ke_list[m],   dtype=float)
    b = np.asarray(bias_list[m], dtype=float)
    mask = (~np.isnan(k)) & (~np.isnan(b))
    if mask.sum() >= 2:
        cov_Ke_bias[m] = np.cov(k[mask], b[mask], bias=True)[0, 1]

# ==============================================================
# 6) NETWORK-LEVEL degrees + Derrida (safe)
# ==============================================================
mean_degree = np.array(
    [np.mean(np.asarray(degrees_genes[m], dtype=float)) if len(degrees_genes[m]) > 0 else np.nan
     for m in range(N)],
    dtype=float
)

Ke_p       = mean_Ke * mean_bias
Ke_p_plusC = Ke_p + cov_Ke_bias
K_p        = mean_degree * mean_bias

derrida_vals = derrida_values_safe(Fs, Is, degrees, max_degree=max_degree_used, nsim=2000)

# ==============================================================
# 7) MOTIFS: FBL + null2 + FFL + enrichment
# ==============================================================
ad.n_variables = n_variables
ad.N = N

# ---- FBLs (you verified this signature/order) ----
(
    nr_loops,
    nr_pos_loops,          # (6, N)
    nr_neg_loops,          # (6, N)
    nr_unknown_loops,
    nr_notreal_loops,
    nr_specific_k_loops,   # (6, 7, N)
    nr_real_loops,
    all_types,
    all_loops
) = ad.compute_all_FBLs(Fs, Is, degrees, constantss, max_degree=max_degree_used)

nr_pos_loops        = np.asarray(nr_pos_loops, dtype=float)
nr_neg_loops        = np.asarray(nr_neg_loops, dtype=float)
nr_specific_k_loops = np.asarray(nr_specific_k_loops, dtype=float)

# totals per network
obs_pos_fbl_total = np.nansum(nr_pos_loops, axis=0)  # (N,)
obs_neg_fbl_total = np.nansum(nr_neg_loops, axis=0)  # (N,)

(
    prop_pos_separate2,               # length N (this is what we need for FFL null)
    prop_pos_separate_by_scc,         # ragged list/object (ignore)
    prop_pos_specific_scc_of_loop,    # (6, 7, N)
    prop_pos_specific_scc_of_loop2
) = ad.get_null_expectations_fbl(Fs, Is, variabless, all_types, all_loops)

prop_pos_specific_scc_of_loop = np.asarray(prop_pos_specific_scc_of_loop, dtype=float)

exp_pos_fbl_null2 = np.nansum(nr_specific_k_loops * prop_pos_specific_scc_of_loop, axis=(0, 1))  # (N,)
total_real_fbl    = np.nansum(nr_specific_k_loops, axis=(0, 1))                                   # (N,)
exp_neg_fbl_null2 = total_real_fbl - exp_pos_fbl_null2

FBL_pos_over_exp_null2 = safe_div(obs_pos_fbl_total, exp_pos_fbl_null2)
FBL_neg_over_exp_null2 = safe_div(obs_neg_fbl_total, exp_neg_fbl_null2)

# ---- FFLs ----
# Signature: compute_all_FFLs(Fs, Is, degrees, constantss, max_degree, N)
ffl_res = ad.compute_all_FFLs(Fs, Is, degrees, constantss, max_degree_used, N)

# take first 4 outputs (ignore any extras)
nr_coherent_ffls   = np.asarray(ffl_res[0], dtype=float)
nr_incoherent_ffls = np.asarray(ffl_res[1], dtype=float)
nr_unknown_ffls    = ffl_res[2]
nr_specific_ffls_raw = ffl_res[3]

# Your current observation: nr_specific_ffls_raw is (N,)
# We treat it as TOTAL FFL count per network for the null expectations.
total_ffls = np.asarray(nr_specific_ffls_raw, dtype=float).reshape(-1)
if total_ffls.shape[0] != N:
    # If it's something else (rare), try to coerce list-of-length-N
    try:
        total_ffls = np.asarray(list(nr_specific_ffls_raw), dtype=float).reshape(-1)
    except Exception as e:
        raise ValueError(f"Could not interpret ffl_res[3] as total FFL counts. type={type(nr_specific_ffls_raw)}") from e
if total_ffls.shape[0] != N:
    raise ValueError(f"FFL total counts length mismatch: got {total_ffls.shape[0]} expected {N}")

# ---- Expected FFL counts per type under the same null model core as paper code ----
prop_pos_separate = np.asarray(prop_pos_separate2, dtype=float).reshape(-1)
if prop_pos_separate.shape[0] != N:
    raise ValueError(f"prop_pos_separate length mismatch: got {prop_pos_separate.shape[0]} expected {N}")

expected_ffl_per_type = expected_ffl_per_type_from_total(total_ffls, prop_pos_separate)  # (8, N)

# classify coherent/incoherent types
coherent_type_ids = []
incoherent_type_ids = []
for t in range(8):
    signs = np.array([
        -1 if (t & 1) else  1,
        -1 if (t & 2) else  1,
        -1 if (t & 4) else  1
    ], dtype=int)
    if can.is_ffl_coherent(signs):
        coherent_type_ids.append(t)
    else:
        incoherent_type_ids.append(t)

exp_coherent_ffl   = np.nansum(expected_ffl_per_type[coherent_type_ids, :], axis=0)   # (N,)
exp_incoherent_ffl = np.nansum(expected_ffl_per_type[incoherent_type_ids, :], axis=0) # (N,)

FFL_coherent_over_exp   = safe_div(nr_coherent_ffls, exp_coherent_ffl)
FFL_incoherent_over_exp = safe_div(nr_incoherent_ffls, exp_incoherent_ffl)

# ==============================================================
# 8) Assemble master DataFrame
# ==============================================================
df = pd.DataFrame({
    "model_name":                    models_loaded,

    # network/structural
    "n_genes":                       n_variables.astype(int),
    "n_constants":                   n_constants.astype(int),
    "mean_degree":                   mean_degree,
    "mean_deg_essential":            mean_Ke,
    "cov_Ke_bias":                   cov_Ke_bias,
    "Ke_times_p1_minus_p":           Ke_p,
    "Ke_times_p1_minus_p_plus_cov":  Ke_p_plusC,
    "K_times_p1_minus_p":            K_p,

    # functional
    "mean_depth":                    mean_depth,
    "prop_canalization_fraction":    prop_ncf,
    "mean_canalizing_strength":      mean_can_strength,
    "mean_input_redundancy":         mean_IR,
    "mean_effective_connectivity":   mean_EC,
    "mean_bias_p1_minus_p":          mean_bias,
    "derrida_value":                 derrida_vals,

    # motifs (raw totals per network)
    "FBL_pos_count":                 obs_pos_fbl_total,
    "FBL_neg_count":                 obs_neg_fbl_total,
    "FFL_coherent_count":            nr_coherent_ffls,
    "FFL_incoherent_count":          nr_incoherent_ffls,

    # motifs (enrichment)
    "FBL_pos_over_exp_null2":        FBL_pos_over_exp_null2,
    "FBL_neg_over_exp_null2":        FBL_neg_over_exp_null2,
    "FFL_coherent_over_exp":         FFL_coherent_over_exp,
    "FFL_incoherent_over_exp":       FFL_incoherent_over_exp,
})

# ==============================================================
# 9) Correlation matrices
# ==============================================================
functional_cols = [
    "mean_depth",
    "prop_canalization_fraction",
    "mean_canalizing_strength",
    "mean_input_redundancy",
    "mean_effective_connectivity",
    "mean_bias_p1_minus_p",
    "derrida_value",
]

network_cols = [
    "n_genes",
    "n_constants",
    "mean_degree",
    "mean_deg_essential",
    "cov_Ke_bias",
    "Ke_times_p1_minus_p",
    "Ke_times_p1_minus_p_plus_cov",
    "K_times_p1_minus_p",

    "FBL_pos_count",
    "FBL_neg_count",
    "FBL_pos_over_exp_null2",
    "FBL_neg_over_exp_null2",

    "FFL_coherent_count",
    "FFL_incoherent_count",
    "FFL_coherent_over_exp",
    "FFL_incoherent_over_exp",
]

corr_functional = df[functional_cols].corr(method="pearson")
corr_network    = df[network_cols].corr(method="pearson")

print("\n=== Functional metric correlation matrix ===")
display(corr_functional)

print("\n=== Network / structural (+ motifs) correlation matrix ===")
display(corr_network)

# Save
corr_functional.to_csv("corr_functional_scores.csv")
corr_network.to_csv("corr_network_scores_with_motifs.csv")
df.to_csv("all_scores_per_network.csv", index=False)

# ==============================================================
# 10) Optional: heatmaps (matplotlib only; no seaborn)
# ==============================================================
def plot_corr_heatmap(corr_df, title, out_png=None):
    M = corr_df.values
    labels = list(corr_df.columns)

    plt.figure(figsize=(0.6*len(labels) + 4, 0.6*len(labels) + 4))
    plt.imshow(M, vmin=-1, vmax=1, interpolation="nearest")
    plt.colorbar()
    plt.xticks(range(len(labels)), labels, rotation=90)
    plt.yticks(range(len(labels)), labels)
    plt.title(title)
    plt.tight_layout()
    if out_png is not None:
        plt.savefig(out_png, dpi=200)
    plt.show()

plot_corr_heatmap(corr_functional, "Functional-score correlations", out_png="corr_functional_heatmap.png")
plot_corr_heatmap(corr_network, "Network-score (+ motif) correlations", out_png="corr_network_heatmap.png")

```

# Validation 

```python
# ==============================================================
# SANITY-CHECK + VALIDATION SUITE FOR CORRELATIONS
# Assumes you already have:
#   df              : master DataFrame (one row per network)
#   functional_cols : list of functional metric column names
#   network_cols    : list of network (+ motif) metric column names
#   corr_functional : Pearson corr matrix for functional_cols (optional)
#   corr_network    : Pearson corr matrix for network_cols (optional)
#
# Produces:
#   - basic matrix sanity checks
#   - pairwise scatter plots (with r, spearman)
#   - residualized (size-controlled) correlations
#   - null-model integrity checks (mean enrichment ~1, enrichment vs degree/size)
#   - Pearson vs Spearman comparison
#   - subsample stability (bootstrap-style)
#
# Requirements: numpy, pandas, matplotlib, scipy, sklearn
# ==============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import spearmanr
from sklearn.linear_model import LinearRegression

# -----------------------------
# Helpers
# -----------------------------
def _finite_xy(x, y):
    x = np.asarray(x, dtype=float).reshape(-1)
    y = np.asarray(y, dtype=float).reshape(-1)
    m = np.isfinite(x) & np.isfinite(y)
    return x[m], y[m], m.sum()

def pearson_r(x, y):
    x, y, n = _finite_xy(x, y)
    if n < 3:
        return np.nan, n
    r = np.corrcoef(x, y)[0, 1]
    return float(r), int(n)

def spearman_r(x, y):
    x, y, n = _finite_xy(x, y)
    if n < 3:
        return np.nan, n
    r, p = spearmanr(x, y)
    return float(r), int(n)

def residualize(y, X):
    """
    Residualize y against X using linear regression.
    y: (N,)
    X: (N,k)
    """
    y = np.asarray(y, dtype=float).reshape(-1)
    X = np.asarray(X, dtype=float)
    m = np.isfinite(y) & np.all(np.isfinite(X), axis=1)
    if m.sum() < 5:
        return np.full_like(y, np.nan, dtype=float)
    model = LinearRegression().fit(X[m], y[m])
    yhat = model.predict(X[m])
    res = np.full_like(y, np.nan, dtype=float)
    res[m] = y[m] - yhat
    return res

def scatter_with_stats(df, xcol, ycol, title=None, alpha=0.6):
    x = df[xcol].values
    y = df[ycol].values
    r_p, n_p = pearson_r(x, y)
    r_s, n_s = spearman_r(x, y)

    plt.figure(figsize=(5.5, 4.5))
    plt.scatter(x, y, s=18, alpha=alpha)
    plt.xlabel(xcol)
    plt.ylabel(ycol)
    if title is None:
        title = f"{ycol} vs {xcol}"
    plt.title(f"{title}\nPearson r={r_p:.3f} (n={n_p}), Spearman ={r_s:.3f} (n={n_s})")
    plt.tight_layout()
    plt.show()

def check_corr_matrix(C, name="corr"):
    C = np.asarray(C, dtype=float)
    ok_sym = np.allclose(C, C.T, equal_nan=True)
    ok_diag = np.allclose(np.diag(C), 1.0, equal_nan=False)
    print(f"[{name}] symmetric:", ok_sym)
    print(f"[{name}] diag==1:", ok_diag)

def top_abs_corr_pairs(df, cols, k=15, min_n=30):
    """
    Show top absolute Pearson correlations among cols (excluding diagonal).
    """
    M = df[cols].corr(method="pearson")
    pairs = []
    for i in range(len(cols)):
        for j in range(i+1, len(cols)):
            a, b = cols[i], cols[j]
            r = M.loc[a, b]
            n = _finite_xy(df[a].values, df[b].values)[2]
            if np.isfinite(r) and n >= min_n:
                pairs.append((abs(r), r, n, a, b))
    pairs.sort(reverse=True, key=lambda t: t[0])
    out = pairs[:k]
    print(f"Top {k} |r| pairs (Pearson), min_n={min_n}:")
    for ar, r, n, a, b in out:
        print(f"  r={r:+.3f} (n={n})  {a}  vs  {b}")
    return out

def subsample_corr_stability(df, cols, n_iter=200, frac=0.8, seed=0):
    """
    Subsample rows and recompute correlation matrices; report sign-flip rates and std of r.
    """
    rng = np.random.default_rng(seed)
    cols = list(cols)
    N = df.shape[0]

    # store correlations for each pair across iterations
    idx_pairs = []
    for i in range(len(cols)):
        for j in range(i+1, len(cols)):
            idx_pairs.append((i, j))

    Rs = { (i,j): [] for (i,j) in idx_pairs }

    for t in range(n_iter):
        take = rng.choice(N, size=int(frac*N), replace=False)
        sub = df.iloc[take][cols]
        C = sub.corr(method="pearson").values
        for (i,j) in idx_pairs:
            r = C[i, j]
            if np.isfinite(r):
                Rs[(i,j)].append(r)

    # summarize
    summary = []
    for (i,j), arr in Rs.items():
        arr = np.asarray(arr, dtype=float)
        if arr.size < max(20, 0.2*n_iter):
            continue
        mean_r = float(np.nanmean(arr))
        std_r  = float(np.nanstd(arr))
        sign_flip = float(np.mean(np.sign(arr) != np.sign(mean_r)))
        summary.append((abs(mean_r), mean_r, std_r, sign_flip, cols[i], cols[j], arr.size))

    summary.sort(reverse=True, key=lambda t: t[0])
    print(f"Subsample stability over {n_iter} iters (frac={frac}):")
    print("Showing top 20 by |mean r| with std and sign-flip rate:")
    for ar, mean_r, std_r, flip, a, b, m in summary[:20]:
        print(f"  mean r={mean_r:+.3f}, std={std_r:.3f}, flip={flip:.2f} (m={m})  {a} vs {b}")
    return summary

# ==============================================================
# 0) Basic DataFrame checks
# ==============================================================
print("df shape:", df.shape)
print("missingness (fraction NaN) for key cols:")
for c in ["n_genes", "mean_degree", "mean_bias_p1_minus_p",
          "FBL_pos_over_exp_null2", "FFL_coherent_over_exp"]:
    if c in df.columns:
        print(f"  {c:28s} {df[c].isna().mean():.3f}")

# ==============================================================
# 1) Correlation matrix sanity checks (symmetry + diag)
# ==============================================================
# If you already computed corr_network/corr_functional earlier, you can use those.
corr_network = df[network_cols].corr(method="pearson")
corr_functional = df[functional_cols].corr(method="pearson")

check_corr_matrix(corr_network.values, name="corr_network (Pearson)")
check_corr_matrix(corr_functional.values, name="corr_functional (Pearson)")

# ==============================================================
# 2) Top correlations (quick scan for redundancies / surprises)
# ==============================================================
_ = top_abs_corr_pairs(df, network_cols, k=15, min_n=30)
_ = top_abs_corr_pairs(df, functional_cols, k=15, min_n=30)

# ==============================================================
# 3) Mandatory scatter sanity plots (edit list as needed)
# ==============================================================
pairs_to_check = [
    ("mean_degree", "FFL_coherent_count"),
    ("mean_degree", "FBL_pos_count"),
    ("Ke_times_p1_minus_p", "mean_canalizing_strength"),
    ("FFL_coherent_count", "FFL_coherent_over_exp"),
    ("FBL_pos_count", "FBL_pos_over_exp_null2"),
    ("mean_bias_p1_minus_p", "Ke_times_p1_minus_p"),
]

for xcol, ycol in pairs_to_check:
    if xcol in df.columns and ycol in df.columns:
        scatter_with_stats(df, xcol, ycol)

# ==============================================================
# 4) Size-controlled (residualized) correlations
#    Controls: n_genes (optionally also n_constants)
# ==============================================================
controls_1 = ["n_genes"]
controls_2 = ["n_genes", "n_constants"]

def residual_corr(df, xcol, ycol, control_cols):
    X = df[control_cols].values
    rx = residualize(df[xcol].values, X)
    ry = residualize(df[ycol].values, X)
    r, n = pearson_r(rx, ry)
    return r, n

size_tests = [
    ("mean_degree", "FFL_coherent_count"),
    ("mean_degree", "FBL_pos_count"),
    ("FFL_coherent_count", "FFL_coherent_over_exp"),
    ("FBL_pos_count", "FBL_pos_over_exp_null2"),
    ("Ke_times_p1_minus_p", "mean_canalizing_strength"),
]

print("\n=== Residualized (size-controlled) Pearson correlations ===")
for xcol, ycol in size_tests:
    if xcol in df.columns and ycol in df.columns:
        r1, n1 = residual_corr(df, xcol, ycol, controls_1)
        r2, n2 = residual_corr(df, xcol, ycol, controls_2)
        print(f"{ycol} vs {xcol:24s} | control {controls_1}: r={r1:+.3f} (n={n1}) | control {controls_2}: r={r2:+.3f} (n={n2})")

# ==============================================================
# 5) Null-model integrity checks
#    - mean enrichment ~ 1
#    - enrichment should not correlate strongly with degree/size
# ==============================================================
enrichment_cols = [c for c in df.columns if c.endswith("_over_exp_null2") or c.endswith("_over_exp")]
print("\n=== Enrichment mean checks (should be ~1; allow modest deviation) ===")
for c in enrichment_cols:
    mu = np.nanmean(df[c].values.astype(float))
    med = np.nanmedian(df[c].values.astype(float))
    print(f"  {c:28s} mean={mu:.3f} median={med:.3f}")

print("\n=== Enrichment vs degree/size (Pearson) ===")
for c in enrichment_cols:
    for base in ["mean_degree", "n_genes", "n_constants"]:
        if base in df.columns:
            r, n = pearson_r(df[c].values, df[base].values)
            print(f"  {c:28s} vs {base:12s} r={r:+.3f} (n={n})")

# Scatter enrichment vs degree (quick visual)
for c in enrichment_cols:
    if "mean_degree" in df.columns:
        scatter_with_stats(df, "mean_degree", c, title=f"{c} vs mean_degree")

# ==============================================================
# 6) Pearson vs Spearman heatmaps (diagnose outliers/nonlinearity)
# ==============================================================
corr_network_s = df[network_cols].corr(method="spearman")
corr_functional_s = df[functional_cols].corr(method="spearman")

def plot_corr_heatmap(corr_df, title):
    M = corr_df.values
    labels = list(corr_df.columns)
    plt.figure(figsize=(0.6*len(labels) + 4, 0.6*len(labels) + 4))
    plt.imshow(M, vmin=-1, vmax=1, interpolation="nearest")
    plt.colorbar()
    plt.xticks(range(len(labels)), labels, rotation=90)
    plt.yticks(range(len(labels)), labels)
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_corr_heatmap(corr_network, "Network correlations (Pearson)")
plot_corr_heatmap(corr_network_s, "Network correlations (Spearman)")

# Optional: show difference matrix
plot_corr_heatmap(corr_network - corr_network_s, "Pearson - Spearman (Network)")

# ==============================================================
# 7) Subsample stability (detect sign flips / instability)
# ==============================================================
_ = subsample_corr_stability(df, network_cols, n_iter=200, frac=0.8, seed=0)

# ==============================================================
# 8) Derrida consistency checks (dynamic sanity)
# ==============================================================
if "derrida_value" in df.columns:
    derr_tests = [
        ("Ke_times_p1_minus_p", "derrida_value"),
        ("mean_degree", "derrida_value"),
        ("FFL_coherent_over_exp", "derrida_value"),
        ("FBL_pos_over_exp_null2", "derrida_value"),
    ]
    print("\n=== Derrida sanity correlations ===")
    for xcol, ycol in derr_tests:
        if xcol in df.columns:
            r, n = pearson_r(df[xcol].values, df[ycol].values)
            rs, ns = spearman_r(df[xcol].values, df[ycol].values)
            print(f"  {ycol} vs {xcol:24s} Pearson r={r:+.3f} (n={n}) | Spearman ={rs:+.3f} (n={ns})")
            scatter_with_stats(df, xcol, ycol)

print("\nDONE: sanity checks complete.")
```
